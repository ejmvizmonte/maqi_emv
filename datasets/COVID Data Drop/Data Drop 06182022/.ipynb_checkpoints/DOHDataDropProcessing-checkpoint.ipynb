```python
import pandas as pd
import numpy as np
import epyestim
import epyestim.covid19 as covid19
```

```python
ddrop1 = pd.read_csv("DOH COVID Data Drop_ 20220611 - 04 Case Information_batch_0.csv", dtype={"CityMuniPSGC":str, "BarangayRes":str, "BarangayPSGC":str})
```

```python
ddrop2 = pd.read_csv("DOH COVID Data Drop_ 20220611 - 04 Case Information_batch_1.csv", dtype={"CityMuniPSGC":str, "BarangayRes":str, "BarangayPSGC":str})
```

```python
ddrop3 = pd.read_csv("DOH COVID Data Drop_ 20220611 - 04 Case Information_batch_2.csv", dtype={"CityMuniPSGC":str, "BarangayRes":str, "BarangayPSGC":str})
```

```python
ddrop = [ddrop1, ddrop2, ddrop3]
```

```python
conso = pd.concat(ddrop, ignore_index=True)
```

```python
conso
```

## Regional Segmentation
NCR

Region IV-A: CALABARZON

BARMM


### NCR

```python
ncr = conso[conso["RegionRes"] == "NCR"]
```

```python
ncr_sample = ncr[(ncr['DateRepConf'] > '2020-01-01') & (ncr['DateRepConf'] < '2022-04-01')]
```

```python
ncr_sample
```

```python
ncrsumm = pd.pivot_table(ncr_sample, index='DateRepConf', values='CaseCode', aggfunc='count')
```

```python
ncrsumm
```

```python
ncrsumm.to_csv("ncrcasecount.csv")
```
